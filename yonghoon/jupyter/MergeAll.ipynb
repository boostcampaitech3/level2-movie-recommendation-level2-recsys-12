```python
import pandas as pd
import numpy as np

```

```python
# 데이터 읽어 오기 
user_cols = ['userid', 'itemid', 'timestamp']
users = pd.read_csv('../data/train/train_ratings.csv', sep=',')

directors = pd.read_csv('../data/train/directors.tsv', sep='\t', encoding='latin-1')
directors = directors.groupby('item', as_index = False, sort = True)['director'].apply(lambda x: "{%s}" % '|'.join(x))

genres = pd.read_csv('../data/train/genres.tsv', sep='\t',encoding='latin-1')
genres = genres.groupby('item', as_index = False, sort = True)['genre'].apply(lambda x: "{%s}" % '|'.join(x))

titles = pd.read_csv('../data/train/titles.tsv', sep='\t',encoding='latin-1')
titles = titles.sort_values(by=['item'])

writers = pd.read_csv('../data/train/writers.tsv', sep='\t', encoding='latin-1')
writers = writers.groupby('item', as_index = False, sort = True)['writer'].apply(lambda x: "{%s}" % '|'.join(x))

years = pd.read_csv('../data/train/years.tsv', sep='\t',encoding='latin-1')
years = years.sort_values(by=['item'])
```

```python
titles
```

```python
# movie 다 합쳐 보자
movies = pd.merge(titles, directors, left_on = 'item', right_on ='item', how = 'left')
movies = pd.merge(movies, writers, left_on = 'item', right_on = 'item', how = 'left')
movies = pd.merge(movies, years, left_on = 'item', right_on = 'item', how = 'left')
movies = pd.merge(movies, genres, left_on = 'item', right_on = 'item', how = 'left')
```

```python
movies.to_csv('../data/train/merge.csv')
```
