```python
# timestamp 제거 
users = users.drop('time', axis=1)
users['rating'] = 1
```

```python
movie_user_saw = users.pivot_table('rating', index='item', columns='user')
user_movie_saw = users.pivot_table('rating', index='user', columns='item')
```

```python
movie_user_saw.head()
```

```python
movie_user_saw.fillna(0, inplace = True)
movie_user_saw.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_based_collabor = cosine_similarity(movie_user_saw)
item_based_collabor
```

```python
item_based_collabor = pd.DataFrame(data = item_based_collabor, index = movie_user_saw.index, columns=movie_user_saw.index)
item_based_collabor
```

```python
rating_df = pd.read_csv('../data/train/train_ratings.csv')
users = rating_df["user"].unique()

result = []

for i, user in enumerate(users):
    if i % 100 == 0:
        print(str(i) + "done")
    is_saw = rating_df['user'] == user
    movies = rating_df[is_saw]['item'].tolist()

    items = item_based_collabor[~item_based_collabor.columns.isin(movies)].sum(axis=1).sort_values(ascending=False)[:10]

    for i in range(10):
        result.append((user, items.index[i]))
    
```

```python
pd.DataFrame(result, columns=["user", "item"]).to_csv(
    "../output/cosine_sim_submission2.csv", index=False
)
```
