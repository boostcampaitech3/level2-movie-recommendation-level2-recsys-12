```python
topk = 9
```

```python
result = []
backup = 0
backup_index = 0
search_list = []
for i in range(len(sequential)):
    
    user = sequential['user'][i]
    if i % 10 == 0 :
        search_list.clear()
        split_static = static['item'][i:i+10].to_list()
        split_sequential = sequential['item'][i:i+10].to_list()
        
        split_sequential = [a for a in split_sequential if a not in split_static[:topk]]
        endpoint = topk
        
        if len(split_sequential) < (10 - topk) :
            endpoint = 10 - len(split_sequential)
            
        for k in range(endpoint) :
            result.append((user, split_static[k]))
        for j in range(10-endpoint) :
            result.append((user, split_sequential[j]))
```

```python
pd.DataFrame(result, columns=["user", "item"]).to_csv(
    os.path.join('./Best + NMF.csv'), index=False
)
```
