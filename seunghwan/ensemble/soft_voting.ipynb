```python
import pandas as pd
import numpy as np
```

```python
bert_data = pd.DataFrame(columns=['user', 'item', 'score'],dtype=object)
for i in range(5):
    temp = pd.read_csv(f'voting_data/burt4Rec_1000history_{i}.csv')
    bert_data = pd.concat([bert_data, temp])
```

```python
recvae_data = pd.read_csv('voting_data/RecVAE.csv')
```

```python
bert_data.reset_index(inplace=True, drop=True)
```

```python
recvae_data.drop(recvae_data[recvae_data['score'] == -np.inf].index, inplace=True)
```

```python
recvae_data['norm_score'] = recvae_data.groupby('user')['score'].transform(lambda x: (x - x.mean()) / x.std())
bert_data['norm_score'] = bert_data.groupby('user')['score'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
recvae_data.drop('score', axis=1, inplace=True)
bert_data.drop('score', axis=1, inplace=True)
```

```python
recvae_data
```

```python
merged_df = pd.merge(recvae_data, bert_data, on=['user', 'item'], suffixes=['R', 'B'])
```

```python
merged_df['no_w_sum'] = merged_df['norm_scoreB'] + merged_df['norm_scoreR']
```

```python
merged_df['w_sum'] = 11 * merged_df['norm_scoreR'] + 1 * merged_df['norm_scoreB']
```

```python
merged_df.drop('norm_scoreR', axis=1, inplace=True)
merged_df.drop('norm_scoreB', axis=1, inplace=True)
```

```python
merged_df.drop('no_w_sum', axis=1, inplace=True)
```

```python
merged_df = merged_df.sort_values('w_sum', ascending=False)
```

```python
merged_df.groupby('user').head(10).to_csv('./voting_data/temp.csv', index=False)
```
