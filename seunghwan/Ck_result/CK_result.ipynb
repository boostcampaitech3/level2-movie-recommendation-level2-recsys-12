```python
import pandas as pd
```

```python
rating = pd.read_csv("/opt/ml/input/data/train/train_ratings.csv")
title = pd.read_csv("/opt/ml/input/data/train/titles.tsv", sep='\t')
result = pd.read_csv("./Bert4rec+RecVAE1.csv")
```

```python
title_dic = dict()
for item, title in zip(title['item'], title['title']) :
    title_dic[item] = title
```

```python
rating['item'] = rating.item.map(title_dic)
```

```python
rating
```

```python
result['item'] = result.item.map(title_dic)
```

```python
result
```

```python
def view_title(id) :
    id = list(result.user.unique())[id]
    for t in rating[rating['user'] == id]['item']:
        print(t)
    print("===============================")
    for t in result[result['user'] == id]['item']:
        print(t)
```

```python
view_title(23)
```
